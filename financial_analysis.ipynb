```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\nasee\Downloads\financial_data.csv")
```

```python
df
```

```python
df.sort_values(['Company', 'Year'], inplace=True)  # Ensure data is sorted
df['Revenue Growth (%)'] = df.groupby('Company')['Total revenue ($M)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income ($M)'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets ($M)'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total liabilities ($M)'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Ops ($M)'].pct_change() * 100
```

```python
summary = df.groupby('Company')[['Revenue Growth (%)', 'Net Income Growth (%)', 'Assets Growth (%)', 'Liabilities Growth (%)', 'Cash Flow Growth (%)']].mean()
print(summary)
```

Financial Performance Analysis (Apple, Microsoft, Tesla)
1. Revenue Growth Analysis
- Apple: Experienced a slight revenue decline of -0.39%, indicating potential challenges in sales growth. This could be due to market saturation, supply chain disruptions, or weaker consumer demand.
- Microsoft: Achieved 11.28% revenue growth, showcasing strong demand for its cloud services, enterprise solutions, and software products.
- Tesla: Reported 9.87% revenue growth, reflecting continued expansion in EV sales despite industry-wide challenges such as supply chain constraints and increasing competition.
2. Net Income Growth Analysis
- Apple: Net income declined by -3.09%, suggesting that increased costs, lower revenue, or margin pressure affected profitability.
- Microsoft: Posted 10.64% growth in net income, indicating strong operational efficiency and high-margin business growth, particularly in cloud computing.
- Tesla: Faced a significant -16.63% drop in net income, which might be attributed to higher production costs, price cuts, or margin pressures in the EV sector.
3. Total Assets Growth Analysis
- Apple: Grew total assets by 1.73%, showing modest expansion in its asset base, potentially through investments in R&D, acquisitions, or infrastructure.
- Microsoft: Witnessed a 18.62% increase in assets, possibly due to strategic investments in AI, cloud computing, and acquisitions.
- Tesla: Reported 21.99% asset growth, the highest among the three companies, reflecting continued expansion in factories, production capacity, and technological advancements.
4. Total Liabilities Growth Analysis
- Apple: Liabilities grew by 1.10%, indicating stable debt management without significant borrowing.
- Microsoft: Saw 11.10% growth in liabilities, likely reflecting investment-driven borrowing to support expansion in AI and cloud services.
- Tesla: Experienced a 15.27% increase in liabilities, which may be due to capital investments in manufacturing, supply chain costs, or new debt issuance to fund growth.
5. Cash Flow from Operating Activities Growth
- Apple: Declined by -1.14%, suggesting slightly weaker cash generation from core operations, possibly due to changes in working capital or revenue stagnation.
- Microsoft: Increased 16.86%, highlighting strong operational cash flow, likely driven by recurring revenue streams from cloud and enterprise services.
- Tesla: Grew 1.30%, indicating stable cash flow despite profitability challenges, likely due to effective cost control and operational efficiency.
6. Key Takeaways
- Microsoft is the strongest performer, with double-digit growth across revenue, income, and assets, driven by cloud and AI investments.
- Apple shows stability but faces challenges, with slight revenue decline and negative income growth, indicating potential margin pressures.
- Tesla is expanding aggressively, but the drop in net income suggests profitability challenges, likely from higher costs or price reductions.
